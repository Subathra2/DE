```python
import pandas as pd
import os
import sqlite3
import numpy as np 
from lets_plot import*
LetsPlot.setup_html()
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename="D:/me204/Meesho Project/Data/Clean/Meesho_Project_Data.csv"
df=pd.read_csv(filename)
```

```python
df
```

```python
conn = sqlite3.connect(os.path.join('../data', 'meesho.db'))
```

```python
df.to_sql('Meesho_products' ,conn, if_exists='replace', index=False)
```

```python
meesho_product=conn.execute('''
    select * from Meesho_products
''').fetchall()
```

```python
meesho_df=df = pd.DataFrame(meesho_product, columns=['Names','Sizes','Prices', 'Fabrics', 'Sleeves','Net_quantities','Patterns', 'Ratings','Delivery'])
```

```python
meesho_df.shape
```

```python
meesho_df
```

```python
ggplot (meesho_df, aes(x='Fabrics',y='Sizes')) + \
    geom_bar(star='identify',fill='green') + \
    ggtitle('Number of fabrics by Ratings') + \
    xlab('Fabrics') + \
    ylab('Ratings') + \
    theme(axis_text_x=element_text(angle=45,hjust=1)
    )
```

```python
# Plot hexbin
plt.figure(figsize=(8, 6))
plt.hexbin(df['Ratings'], df['Prices'], gridsize=30, cmap='coolwarm', mincnt=1)
plt.colorbar(label="Count of Products")
plt.xlabel("Ratings")
plt.ylabel("Prices")
plt.title("Prices vs Rating")
plt.show()
```

```python
(
    ggplot(data = meesho_df, mapping = aes(
                    x = 'Prices', 
                    y = 'Fabrics', 
                    color = 'Prices', 
                    size = 'Sizes')
             ) +
        geom_point(alpha=0.5) +
        scale_x_log10() +
        theme_grey()        
)
```

```python
(
    ggplot(
        data = meesho_df,
        mapping = aes(
            x = 'Patterns',
            y = 'Ratings',
            color = 'Prices'
            )
        ) +
        geom_point()+
        geom_smooth(method = 'lm') +
        theme_grey()

)
```

```python
df = pd.DataFrame(meesho_df)

# Calculate the count of each rating value
fabrics_counts = df['Fabrics'].value_counts()

# Plot pie chart
plt.figure(figsize=(10,8))
plt.pie(fabrics_counts, labels=fabrics_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Product fabric')
plt.show()
```

```python
df = pd.DataFrame(meesho_df)

# Calculate the count of each rating value
rating_counts = df['Sleeves'].value_counts()

# Plot bar chart
plt.figure(figsize=(10,10))
rating_counts.sort_index().plot(kind='line', color='brown')
plt.title('Distribution of Product sleeves')
plt.xlabel('Sleeves')
# plt.ylabel('prices')
# plt.xticks(rotation=0)  # Keeps the x-axis labels horizontal
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Prices"], bins=30, color="purple", edgecolor="black")
plt.xlabel("Prices")
plt.ylabel("size")
plt.title("Price Distribution")
plt.show()
```
